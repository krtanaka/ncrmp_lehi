```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
```

```python
#lon_range = c(143, 207)
#lat_range = c(-16, 30)

lon_min = 143.
lon_max = 207.
lat_min = -16.
lat_max = 30.
```

```python
url_dhw = 'https://oceanwatch.pifsc.noaa.gov/erddap/griddap/CRW_dhw_v1_0'
dhw_ds = xr.open_dataset(url_dhw)
dhw_ds
```

```python
dhw_subset = dhw_ds['degree_heating_week'].sel(
                                     latitude=slice(lat_max, lat_min),
                                     longitude=slice(lon_min, lon_max)
                                     )
dhw_subset
```

```python
mon_field=[]
for yr in range(1985,2025):
    
    st_yr = str(yr)+'-01-01'
    end_yr = str(yr)+'-12-31'
    dhw_temp = dhw_subset.sel(time=slice(st_yr,end_yr))
    #print(dhw_temp)
    mon_temp = dhw_temp.groupby('time.month').mean('time')
    print(mon_temp)
    mon_field.append(mon_temp)
mon_field
```

```python
mon_field[0]
```

```python
#np.array(mon_field[1:-1]).shape
TS=[]

for j in range(0,10):
    TS.append(mon_field[0][j,:,:])
        
for i in range(1,39):
    for j in range(0,12):
        TS.append(mon_field[i][j,:,:])
    
for j in range(0,5):
    TS.append(mon_field[-1][j,:,:])
```

```python
np.array(TS).shape
```

```python
fn='./ACCESS-CM2_r1i1p1f1_zg_1900_to_2100_regridded.nc'
ref = xr.open_dataset(fn)
ref.close()
```

```python
dhw_mon= xr.DataArray(np.array(TS),                                 
          name='sst',             
          coords={'time':ref.time.sel(time=slice('1985-03-01','2024-05-31')),'latitude':dhw_subset.latitude,'longitude':dhw_subset.longitude },    
          dims=('time','latitude','longitude') )
dhw_mon
```

```python
dhw_mon.to_netcdf('./dhw_monthly.nc')
```
